```python
import torch
from torchvision import models

# Define the number of classes in your dataset
num_classes = 27  # Example for ASL letters A-Z

# Load the ResNet-18 model and modify the final layer
model = models.mobilenet_v2()

model.classifier[1] = torch.nn.Linear(model.last_channel, num_classes)  # Adjust final layer

# Load the saved weights
model.load_state_dict(torch.load("D:\ASL_Alphabet_Dataset\Sign_language\SignLanguageToText\\asl_to_text_model_74.pth",map_location=torch.device('cpu')))

# Move the model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Set the model to evaluation mode
model.eval()
```

```python
from PIL import Image
from torchvision import transforms

# Define the same transform used in training
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Ensure the input size matches the model
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load and transform the image
def preprocess_image(image_path):
    image = Image.open(image_path)
    image = transform(image).unsqueeze(0)  # Add batch dimension
    return image
```

```python
# Example image path
image_path = r"D:\ASL_Alphabet_Dataset\s.jpg"

# Preprocess the image
image = preprocess_image(image_path)

# Move the image to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
image = image.to(device)

# Make a prediction
with torch.no_grad():  # Disable gradient calculation for inference
    output = model(image)
    _, predicted_class = torch.max(output, 1)  # Get the index of the max log-probability

# Print the predicted class
print(f'Predicted class: {predicted_class.item()}')

```
